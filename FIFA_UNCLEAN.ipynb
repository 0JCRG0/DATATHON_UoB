## BYE REGEX

```python
# replace the newline characters with empty strings
df = df.replace('\n', '', regex=True)
```

```python
df.head()
```

## HEIGHT TO METERS

```python
# define a function to convert height strings to meters
def height_to_meters(height_string):
    feet, inches = height_string.split("'")
    feet = int(feet)
    inches = int(inches.strip('"'))
    total_inches = feet * 12 + inches
    return round(total_inches * 0.0254, 2)

# apply the function to the 'Height' column and assign the result to a new column 'Height (m)'
df['Height (m)'] = df['Height'].apply(height_to_meters)


# drop column 'Height'
df = df.drop('Height', axis=1)

df.head()
```

## MOVE COLUMN TO CERTAIN INDEX

```python
# move 'ID' column to be the first column
id_col = df.pop('ID')
df.insert(0, 'ID', id_col)

# print the updated dataframe
df.head()
```
